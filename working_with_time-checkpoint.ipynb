```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://gist.githubusercontent.com/ryanorsinger/b309f8db19e0ca71b213d4877d835e77/raw/f5841017310e2f4ca070b313529ceec2375336ba/coffee_consumption.csv"
df = pd.read_csv(url)
print(df.dtypes)
df.head()
```

```python
df['date'] = pd.to_datetime(df.date)
df.dtypes
```

# .dt

```python
df['date'].dt.day
```

```python
df['day_of_week'] = df['date'].dt.day_name()
```

```python
df
```

```python
df = df.set_index('date')
```

# asfreq, resample

```python
df.asfreq('M') # D = day, W = week, M = month
```

```python
df.resample('M').sum() #aggregate of coffee_consumption per month
```

```python
df.resample('W').mean()
# you can put in a number followed by letter, e.g. 4D, and it will aggregate every 4 days

```

# forward and back fill

```python
fill_df = df.asfreq('D')
fill_df['forward'] = fill_df['coffee_consumption'].ffill()
```

```python
fill_df['backward'] = fill_df['coffee_consumption'].bfill()
```

```python
fill_df.head(10)
```

```python
fill_df['day_of_week'] = fill_df.index.day_name()
fill_df.head(10)
```

```python
fill_df['coffee_consumption'] = fill_df['coffee_consumption'].fillna(0)
fill_df.head()
```

# plotting

```python
fill_df['coffee_consumption'].plot()
```

```python
fill_df.resample('W').median()['coffee_consumption'].plot()
```

```python
fill_df.resample('M').median()['coffee_consumption'].plot()
```

# rolling

```python
fill_df['rolling_week_avg'] = fill_df['coffee_consumption'].rolling(7).mean()
fill_df.head(10)
```

```python
flights = data.flights_20k() # aquiring data 
```

```python
flights.info()
```

```python
flights = flights.set_index('date') #making the date the index
```

```python
flights
```

# 1. changing the negative delays

```python
length = len(flights['delay'])

for i in range(0, length):
    if flights['delay'].iloc[i] < 0:
        flights = flights['delay'].iloc[i] = 0 #for loop that changes negative delay values to zero
```

```python
flights
```

```python
flights.groupby([flights.index.hour], as_index=True).mean() 
#groups all rows by hour of day and averaging the values
```

# 2.  3 am had the highest average delay
