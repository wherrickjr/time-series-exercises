```python
starships_df = [] #for loop that calls API for every REST in starships column
for i in range(0, 82):
    length = len(starships['starships'].loc[i])
    if len(starships['starships'].loc[i]) == 1:
        data = requests.get(starships['starships'].loc[i][0])
        datajson = data.json()
        starships_df.append(datajson)
    else:
        for z in range(0, length):
                       data = requests.get(starships['starships'].loc[i][z])
                       datajson = data.json()
                       starships_df.append(datajson)

starships_df
```

```python
starships_df = pd.DataFrame(starships_df) #turns list into df
```

# 2 do the same thing but for planets.

```python
#request planets api
response = requests.get('https://swapi.dev/api/planets/')
data = response.json()

#turn 'data' within the json into a dataframe
planet_df = pd.DataFrame(data['results'])
```

```python
while data['next'] != None:
    print(data['next'])
    response = requests.get(data['next'])
    data = response.json()
    planet_df = pd.concat([planet_df, pd.DataFrame(data['results'])], ignore_index=True)
```
